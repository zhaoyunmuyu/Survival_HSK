```python
import os
import time
import random
import logging
from typing import Union, List, Tuple

import pandas as pd
import requests

# =============================================================================
# æ—¥å¿—é…ç½®
# =============================================================================
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler("yjw_image_88end.log"),
        logging.StreamHandler()
    ]
)

# =============================================================================
# å¸¸é‡è®¾ç½®
# =============================================================================
HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/91.0.4472.124 Safari/537.36"
    ),
    "Accept": "image/webp,image/apng,image/*,*/*;q=0.8",
    "Accept-Language": "zh-CN,zh;q=0.9"
}

# =============================================================================
# å·¥å…·å‡½æ•°
# =============================================================================
def create_directory(path: str) -> bool:
    """åˆ›å»ºç›®å½•ï¼Œå¦‚æœå·²å­˜åœ¨åˆ™ä¸åšå¤„ç†ã€‚"""
    try:
        os.makedirs(path, exist_ok=True)
        return True
    except Exception as exc:
        logging.error("åˆ›å»ºç›®å½• %s å¤±è´¥: %s", path, exc)
        return False


def normalize_img_entry(entry: str) -> str:
    """
    å¤„ç†å•æ¡å›¾ç‰‡å­—æ®µï¼š
    - å¦‚æœåŒ…å« '|'ï¼Œå–å·¦ä¾§ URLã€‚
    - å»é™¤é¦–å°¾ç©ºæ ¼ã€‚
    """
    entry = entry.strip()
    if "|" in entry:
        return entry.split("|", 1)[0].strip()
    return entry


def extract_image_plan(row: pd.Series, max_images: int = 1) -> List[Tuple[str, str]]:
    """
    ä»å•æ¡è¯„è®ºä¸­æå–éœ€è¦ä¸‹è½½çš„å›¾ç‰‡ (æœ€å¤š max_images å¼ )ï¼Œ
    è¿”å›åˆ—è¡¨ï¼Œæ¯é¡¹åŒ…å« (img_url, photo_id)ã€‚
    """
    # Parse URLs and IDs, ensuring they are non-empty and cleaned
    raw_urls = [
        normalize_img_entry(part)
        for part in str(row.get("review_imgsrc", "")).split("\n")
        if part and str(part).strip().lower() != "nan"
    ]
    raw_ids = [
        str(part).strip()
        for part in str(row.get("review_photo_id", "")).split(";")
        if part and str(part).strip().lower() != "nan"
    ]

    # Handle mismatched counts
    if len(raw_urls) != len(raw_ids):
        logging.warning(
            "å›¾ç‰‡ URL ä¸å›¾ç‰‡ ID æ•°é‡ä¸ä¸€è‡´ï¼Œreview_id=%sï¼Œurls=%dï¼Œids=%d",
            row.get("review_id", "UNKNOWN"),
            len(raw_urls),
            len(raw_ids)
        )

    # Limit the number of images to the smaller of URLs, IDs, and max_images
    limit = min(max_images, len(raw_urls), len(raw_ids))
    return [(raw_urls[i], raw_ids[i]) for i in range(limit)]


def build_image_path(base_dir: str, photo_id: str, img_url: str) -> str:
    """
    æ ¹æ®å›¾ç‰‡ URL çš„æ‰©å±•åå’Œå›¾ç‰‡ ID æ„å»ºä¿å­˜è·¯å¾„ã€‚
    æ²¡æœ‰æ‰©å±•åæ—¶é»˜è®¤ä½¿ç”¨ .jpgã€‚
    """
    _, ext = os.path.splitext(img_url.split("?")[0])
    ext = ext.lower()
    if ext not in {".jpg", ".jpeg", ".png", ".gif", ".webp", ".bmp"}:
        ext = ".jpg"
    filename = f"{photo_id}{ext}"
    return os.path.join(base_dir, filename)


def download_image(url: str, save_path: str, max_retries: int = 5) -> bool:
    """
    ä¸‹è½½å›¾ç‰‡å¹¶ä¿å­˜åˆ°æŒ‡å®šè·¯å¾„ï¼Œå…·å¤‡é‡è¯•æœºåˆ¶ã€‚
    """
    for attempt in range(max_retries):
        try:
            response = requests.get(
                url,
                timeout=15,
                stream=True,
                headers=HEADERS
            )

            if response.status_code == 200:
                with open(save_path, "wb") as file:
                    for chunk in response.iter_content(1024):
                        file.write(chunk)
                logging.info("æˆåŠŸä¸‹è½½: %s", save_path)
                return True

            if response.status_code in (504, 503, 429):
                wait_time = (attempt + 1) * 2
                logging.warning(
                    "çŠ¶æ€ç  %sï¼Œç­‰å¾… %.1fs åé‡è¯•ï¼ŒURL: %s",
                    response.status_code,
                    wait_time,
                    url
                )
                time.sleep(wait_time)
            else:
                logging.warning(
                    "ä¸‹è½½å¤±è´¥ï¼ŒçŠ¶æ€ç : %sï¼ŒURL: %s",
                    response.status_code,
                    url
                )
                time.sleep(2)

        except Exception as exc:
            wait_time = (attempt + 1) * 1.5
            logging.warning(
                "ä¸‹è½½å°è¯• %d å¤±è´¥: %sï¼Œç­‰å¾… %.1fs åé‡è¯•ï¼ŒURL: %s",
                attempt + 1,
                exc,
                wait_time,
                url
            )
            time.sleep(wait_time)

    logging.error("è¶…è¿‡æœ€å¤§é‡è¯•æ¬¡æ•°ï¼Œä¸‹è½½å¤±è´¥: %s", url)
    return False

# =============================================================================
# ä¸»å¤„ç†æµç¨‹
# =============================================================================
def process_reviews(data_source: Union[str, pd.DataFrame],
                    base_dir: str = "../nvme_data02/lijun/image_yjw/",
                    max_images_per_review: int = 1):
    """
    ä¸»æµç¨‹ï¼š
    1. è¯»å–æ•°æ®ã€‚
    2. ä¸ºæ‰€æœ‰è¯„è®ºä¸‹è½½å‰ max_images_per_review å¼ å›¾ç‰‡ã€‚
    3. æ‰€æœ‰å›¾ç‰‡ä¿å­˜åˆ°åŒä¸€ç›®å½•ä¸‹ï¼Œæ–‡ä»¶åä½¿ç”¨ review_photo_idã€‚
    """
    # Load data
    if isinstance(data_source, pd.DataFrame):
        df = data_source
    elif isinstance(data_source, str):
        df = pd.read_csv(data_source)
    else:
        raise ValueError("data_source å¿…é¡»æ˜¯ DataFrame æˆ– CSV æ–‡ä»¶è·¯å¾„")

    # Ensure base directory exists
    if not create_directory(base_dir):
        logging.error("åŸºç¡€ç›®å½•åˆ›å»ºå¤±è´¥ï¼Œç¨‹åºé€€å‡º")
        return

    total_images = 0
    success_images = 0

    logging.info("å¼€å§‹å¤„ç†ï¼Œå…± %d æ¡è¯„è®º", len(df))

    for idx, row in df.iterrows():
        plan = extract_image_plan(row, max_images=max_images_per_review)

        if not plan:
            continue

        for img_url, photo_id in plan:
            total_images += 1
            save_path = build_image_path(base_dir, photo_id, img_url)

            if os.path.exists(save_path):
                logging.info("æ–‡ä»¶å·²å­˜åœ¨ï¼Œè·³è¿‡: %s", save_path)
                success_images += 1
                continue

            # Random delay to avoid request pattern detection
            time.sleep(random.uniform(0.00001, 0.001))

            if download_image(img_url, save_path):
                success_images += 1

        if idx and idx % 100 == 0:
            logging.info("å·²å¤„ç† %d æ¡è¯„è®ºï¼ŒæˆåŠŸ %d/%d", idx, success_images, total_images)

    logging.info("\n===== å¤„ç†å®Œæˆ =====")
    logging.info("æ€»è¯„è®ºæ•°: %d", len(df))
    logging.info("è¯·æ±‚å›¾ç‰‡æ•°: %d", total_images)
    logging.info("æˆåŠŸä¸‹è½½: %d", success_images)
    if total_images:
        logging.info("æˆåŠŸç‡: %.2f%%", success_images / total_images * 100)
    else:
        logging.info("æœªæ‰¾åˆ°å¯ä¸‹è½½çš„å›¾ç‰‡")


if __name__ == "__main__":
    # ä½¿ç”¨ç¤ºä¾‹ï¼š
    # process_reviews("your_reviews.csv")
    # æˆ–è€…ä¼ å…¥å·²åŠ è½½çš„ DataFrameï¼š
    df = pd.read_csv("æ²¹å°–æ—ºé¤å…è¯„è®ºæ•°æ®.csv")
    df2 = df.iloc[880000: ,:]
    process_reviews(df2)
    pass
```
